```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from spectro.utils import vibronic_emission, vibronic_ls
```

```python
%matplotlib notebook
```

```python
def vibronic_emission_v2(x, amp, e0, hrs, sigma, gamma, e_vib, kt=0, **kw):
    return vibronic_ls(-(x-e0), hrs, sigma, gamma, e_vib, kt=kt, **kw)*amp
```

```python
x = np.linspace(1.8, 2.2, 1000)
hrs = 0.2
kt = 0.025
e_vib = 0.025
sigma = 0.001
gamma = 0.001
e0 = 2

ls = vibronic_ls(-(x-e0), hrs, sigma, gamma, e_vib, kt)
emls  = vibronic_emission(x, 1, e0, hrs, sigma, gamma, e_vib, kt)
em_v2 = vibronic_emission_v2(x, 1, e0, hrs, sigma, gamma, e_vib, kt)

assert np.allclose(ls, emls)
assert np.allclose(ls, em_v2)
```

```python
plt.figure()
plt.plot(x, ls)
plt.plot(x, emls)
plt.plot(x, em_v2)
```
